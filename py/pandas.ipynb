Filtering the dataframes is what a lot of DataScience leads with in general.

Say you want to find out what all rows satisfy the condition of less than 10 InternetUsers

```python
stats = stats_new
Filter = stats.InternetUsers < 2
# filter is a column of true false values
```

```python
stats[Filter]
# this will now print only the rows where the condition mentioned before is true
```

```python
filter2 = stats.BirthRate > 40
```

```python
stats[filter2]
# can we combine both the filters together
```

```python
Fin = filter2 & Filter #& is bitwise, dont use and for series
print(stats[Fin])
```

or while making the filter put all conditions together seperated by an & and not and. and is logical and and & and bitwise logical and
